```python
#Training
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
#Visualizer
import os
import graphviz 
os.environ["PATH"] += os.pathsep + "C:\\Program Files (x86)\\Graphviz2.38\\bin"
tree.plot_tree(clf)
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("Decision Tree") 
```

```python
#Prediction
import numpy as np
import csv
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

df = pd.read_csv("C:\\Users\Cv\\Testing_Smote.csv")

X_test = df.iloc[:,:9].astype(int)
y_test = df.iloc[:,9].astype(int)
y_pred = clf.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))


        
        
```

```python
import sklearn.model_selection as model_selection
from imblearn.over_sampling import SMOTE
import pandas as pd

df = pd.read_csv("C:\\Users\Cv\\Training_Smote.csv")

X = df.iloc[:,:9].astype(int)
y = df.iloc[:,9].astype(int)

# setting up testing and training sets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.25, random_state=27)

sm = SMOTE(random_state=27)
X_train, y_train = sm.fit_sample(X_train, y_train)
X_train
X_train.to_csv('TrainingX.csv')
y_train.to_csv('TrainingY.csv')
X_test.to_csv('TestingX.csv')
y_test.to_csv('TestingY.csv')
```
