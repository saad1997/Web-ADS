```python
import csv
import re
# Using readlines() 
file1 = open('latest_malicious.txt', 'r') 
Lines = file1.readlines() 
  
count = 0

client = "NULL"
time = "NULL"
request_type = "NULL"
Accept = "NULL"
Accept_Encoding = "NULL"
Accept_Language = "NULL"
User_Agent = "NULL"
Cookie = "NULL"
Data = "NULL"
Content_Length = "NULL"
Content_Type = "NULL"
Requset_URL = "NULL"

with open("C:/Users/Cv/Desktop/latest_malicious.csv", 'a') as file:
    writer = csv.writer(file)
    # Strips the newline character 
    for line in Lines: 
        if "#" in line:
            line=line.split("-")
            client = line[-1].replace("client","")
            client = re.sub(r"[\n\t\s]*", "", client)
            time = line[0]
            
            
            
        if "--request" in line:
            request = line.replace("--request","")
            request = line.replace("\\","")
            request_type = re.sub(r"[\n\t\s]*", "", request)
            
        if "Accept: " in line:
            accept = line.replace("--header","")
            accept = accept.replace("\\","")
            Accept = re.sub(r"[\n\t\s]*", "", accept)
            
        if "Accept-Encoding: " in line:
            accept_encoding = line.replace("--header","")
            accept_ecoding = accept_encoding.replace("\\","")
            Accept_Encoding = re.sub(r"[\n\t\s]*", "", accept_ecoding)
            
            
        if "Accept-Language: " in line:
            accept_language = line.replace("--header","")
            accept_language = accept_language.replace("\\","")
            Accept_Language = re.sub(r"[\n\t\s]*", "", accept_language)
            
        if "User-Agent: " in line:
            ug = line.replace("--header","")
            ug = ug.replace("\\","")
            User_Agent = re.sub(r"[\n\t\s]*", "", ug)
            
        if "Cookie: " in line:
            ck = line.replace("--header","")
            ck = ck.replace("\\","")
            Cookie = re.sub(r"[\n\t\s]*", "", ck)
            
        if "--data " in line:
            data = line.replace("--data","")
            data = data.replace("\\","")
            Data = re.sub(r"[\n\t\s]*", "", data)
            
            
        if "Content-Length: " in line:
            cl = line.replace("--header ","")
            cl = cl.replace("\\","")
            Content_Length = re.sub(r"[\n\t\s]*", "", cl)
            
            
        if "Content-Type: " in line:
            ct = line.replace("--header ","")
            ct = ct.replace("\\","")
            Content_Type = re.sub(r"[\n\t\s]*", "", ct)
            
        if str(line).startswith("http://"):
            rt = line.replace("\\","")
            Requset_URL = re.sub(r"[\n\t\s]*", "", rt)
            writer.writerow([client,time,request_type,Accept,Accept_Encoding,Accept_Language,User_Agent,Cookie,Data,Content_Length,Content_Type,Requset_URL])
            client = "NULL"
            time = "NULL"
            request_type = "NULL"
            Accept = "NULL"
            Accept_Encoding = "NULL"
            Accept_Language = "NULL"
            User_Agent = "NULL"
            Cookie = "NULL"
            Data = "NULL"
            Content_Length = "NULL"
            Content_Type = "NULL"
            Requset_URL = "NULL"
       
```
